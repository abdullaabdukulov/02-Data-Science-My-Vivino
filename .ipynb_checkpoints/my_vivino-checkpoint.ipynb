```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
def load_and_clean_wine_dataset(wine):
    wine = wine.dropna()
    wine['type'] = wine['type'].str.replace('white', '0')
    wine['type'] = wine['type'].str.replace('red', '1')
    wine['type'] = wine['type'].astype('int')
    return wine
    
data = pd.read_csv("winequalityN.csv")
wine = load_and_clean_wine_dataset(data)
wine
```

```python
class MachineLaerning:
    def __init__(self, test_size, random_state):
        self.test_size = test_size
        self.random_state = random_state
        
    def Logistic_Regression(self, wine):
        X = wine.iloc[:,1:]
        y = wine.iloc[:,0]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=self.test_size, random_state=self.random_state)
        logg_reg = LogisticRegression()
        logg_reg.fit(X_train, y_train)
        y_pred = logg_reg.predict(X_test)
        wine_r2 = r2_score(y_test, y_pred)
        wine_acc = accuracy_score(y_test, y_pred)
        print(f"R2 score: {wine_r2*100}")
        print(f"Accurancy score: {wine_acc*100}")
        
    def Random_Forest_Classifier(self, wine):
        X = wine.iloc[:,1:]
        y = wine.iloc[:,0]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=self.test_size, random_state=self.random_state)
        rand_f = RandomForestClassifier()
        rand_f.fit(X_train, y_train)
        y_pred = rand_f.predict(X_test)
        wine_r2 = r2_score(y_test, y_pred)
        wine_acc = accuracy_score(y_test, y_pred)
        print(f"R2 score: {wine_r2*100}")
        print(f"Accurancy score: {wine_acc*100}")
        
    def Linear_Discriminant_Analysis(self, wine):
        X = wine.iloc[:,1:]
        y = wine.iloc[:,0]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=self.test_size, random_state=self.random_state)
        discr_analysis = LinearDiscriminantAnalysis()
        discr_analysis.fit(X_train, y_train)
        y_pred = discr_analysis.predict(X_test)
        wine_r2 = r2_score(y_test, y_pred)
        wine_acc = accuracy_score(y_test, y_pred)
        print(f"R2 score: {wine_r2*100}")
        print(f"Accurancy score: {wine_acc*100}")


ML = MachineLaerning(test_size=0.2, random_state=42)

ML.Logistic_Regression(wine)
ML.Random_Forest_Classifier(wine)
ML.Linear_Discriminant_Analysis(wine)
```
